```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate user-user similarity matrix
user_similarity = cosine_similarity(combined_data)
user_similarity_df = pd.DataFrame(user_similarity, index=combined_data.index, columns=combined_data.index)
```

```python
# Calculate item-item similarity matrix
item_similarity = cosine_similarity(combined_data.T)
item_similarity_df = pd.DataFrame(item_similarity, index=combined_data.columns, columns=combined_data.columns)
```

```python
def predict_user_based(user, item):
    similar_users = user_similarity_df[user]
    user_ratings = combined_data[item]
    weighted_sum = sum(similar_users * user_ratings)
    similarity_sum = sum(similar_users)
    return weighted_sum / similarity_sum if similarity_sum != 0 else 0

# Predict rating for a specific user and item
predicted_rating = predict_user_based('A', 'Item3')
print(f'Predicted rating for User A on Item3: {predicted_rating}')
```

```python
def predict_item_based(user, item):
    similar_items = item_similarity_df[item]
    user_ratings = combined_data.loc[user]
    weighted_sum = sum(similar_items * user_ratings)
    similarity_sum = sum(similar_items)
    return weighted_sum / similarity_sum if similarity_sum != 0 else 0

# Predict rating for a specific user and item
predicted_rating = predict_item_based('A', 'Item3')
print(f'Predicted rating for User A on Item3: {predicted_rating}')
```

```python
def recommend_items(user, num_recommendations=5):
    predictions = {item: predict_item_based(user, item) for item in combined_data.columns}
    recommended_items = sorted(predictions, key=predictions.get, reverse=True)[:num_recommendations]
    return recommended_items

# Get top 5 recommendations for User A
recommendations = recommend_items('A')
print(f'Top recommendations for User A: {recommendations}')
```
